# Задание №1

```python
import pandas as pd
```

```python
movie_df = pd.read_csv("ml-latest-small/movies.csv")
rating_df = pd.read_csv("ml-latest-small/ratings.csv")
merged_df = movie_df.merge(rating_df, on="movieId", how="left")
merged_df.head()
```

```python
def classify(rating):
    if rating <= 2.0:
        return "низкий рейтинг"
    elif rating <= 4.0:
        return "средний рейтинг"
    else:
        return "высокий рейтинг"
```

```python
merged_df["class"] = merged_df["rating"].apply(classify)
```

# Задание №2

```python
df = pd.read_csv("ml-latest-small/keywords.csv")
df.head()
```

```python
def classify_geo(df):
    """Можно классифицировать по столбцу, но для разнообразия сделал по строке/df"""
    
    geo_data = {
        'Центр': ['москва', 'тула', 'ярославль'],
        'Северо-Запад': ['петербург', 'псков', 'мурманск'],
        'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }

    # Разбиваем запрос на отдельные слова и преобразуем в set,
    # чтобы находить пересечение с множествами городов
    request_set = set(df["keyword"].split())
    
    if request_set & set(geo_data["Центр"]):
        return "Центр"
    elif request_set & set(geo_data["Северо-Запад"]):
        return "Северо-Запад"
    elif request_set & set(geo_data["Дальний Восток"]):
        return "Дальний Восток"
    else:
        return "undefined"
```

```python
df["region"] = df.apply(classify_geo, axis=1)
```
